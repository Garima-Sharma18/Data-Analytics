## Chart Requirement

### **Total Sales by Fat Content**

```python
sales_by_fat_content = df.groupby("Item Fat Content")["Sales"].sum()
plt.pie(sales_by_fat_content, labels=sales_by_fat_content.index, colors=['#fee041', '#0a8421'], autopct='%.1f%%', startangle=90)
plt.title("Total Sales by Fat Content")
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

### **Total Sales by Item type**

```python
sales_by_item_type = df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_item_type.index, sales_by_item_type.values, color=['#fee041', "#C3DFE0", '#0a8421'])

plt.title("Total Sales by Item Type")
plt.xlabel("Item Type")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=-90)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"${yval:,.0f}", ha='center', va='bottom', fontsize=6)

plt.tight_layout()
plt.show()
```

### **Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(["Outlet Location Type", "Item Fat Content"])["Sales"].sum().unstack()
grouped = grouped[["Regular", "Low Fat"]]

ax = grouped.plot(kind='bar', figsize=(8,5), color=['#fee041', '#0a8421'])
plt.title("Outlet Tier by Fat Content")
plt.xlabel("Outlet Location Type")
plt.ylabel("Total Sales ($)")
plt.legend(title="Item Fat Content")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby("Outlet Establishment Year")["Sales"].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle="-", color='#0a8421')

plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales ($)")
plt.title("Sales by Outlet Establishment")

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f"${y:,.0f}", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()  
```

### **Sales by Outlet Size**

```python
sale_by_size = df.groupby(["Outlet Size"])["Sales"].sum()

plt.figure(figsize=(4,4))
plt.pie(sale_by_size, labels=sale_by_size.index, colors=["#C3DFE0", '#fee041', '#0a8421'], autopct='%.1f%%', startangle=90)
plt.title("Sales by Outlet Size")
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

### **Sales by Outlet Location**

```python
sales_by_location = df.groupby("Outlet Location Type")["Sales"].sum().reset_index()
sales_by_location = sales_by_location.sort_values(by="Sales", ascending=False)

plt.figure(figsize=(8, 3))
sns.barplot(x="Sales", y="Outlet Location Type", data=sales_by_location, color='#0a8421')

plt.title("Total Sales by Outlet Location")
plt.xlabel("Total Sales ($)")
plt.ylabel("Outlet Location Type")

plt.tight_layout()
plt.show()
```
